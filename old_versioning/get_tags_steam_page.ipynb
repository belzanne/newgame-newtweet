```python
import requests
from bs4 import BeautifulSoup
import json

def get_game_tags_by_scraping(app_id):
    url = f"https://store.steampowered.com/app/{app_id}/"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        tag_elements = soup.find_all('a', class_='app_tag')
        tags = [tag.text.strip() for tag in tag_elements]
        return tags
    else:
        return f"Erreur lors de la requête : {response.status_code}"

def display_game_tags(app_id):
    tags = get_game_tags_by_scraping(app_id)
    if isinstance(tags, list):
        print(f"\nTags pour le jeu {app_id}:")
        for tag in tags:
            print(f"- {tag}")
    else:
        print(tags)

# ID du jeu HEART ABYSS
app_id = 3119560
display_game_tags(app_id)

# Afficher les données brutes pour une inspection plus détaillée
print("\nDonnées brutes des tags:")
print(json.dumps(get_game_tags_by_scraping(app_id), indent=2))
```
